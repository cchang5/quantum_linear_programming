## Solutions

```python
graph = {(0, 1), (1, 2), (2, 1)}
qubo = get_mds_qubo(graph, directed=True)
e, nodes = classical_search(qubo, n_nodes=3)
print("Brute force search classical minimal energy", e)
print(nodes)
for node in nodes:
    gt.get_plot(graph, node, directed=True)
plt.show()
```
